## Data manipulation

````r
options(repr.matrix.max.rows = 10, repr.matrix.max.cols = 1000)
````

````r
mk_mat <- function(sample_id, is_positive) {
    q <- sprintf("select * from k562_sum_substs_by('%s', '%s')", sample_id, is_positive)
    res <- dbGetQuery(con, q) %>% as_tibble()
    res.spread <- res %>%
        spread(substitution_position, propsum, fill = 0) %>%
        arrange(substitution_value) %>%
        select(!c(substitution_value))
    res.mat <- as.matrix(res.spread, dimnames = c(c("A","C","G","T"), NULL))
    res.mat
}
````

````r
q <- sprintf("select * from map_label")
res <- dbGetQuery(con, q) %>%
    as_tibble() %>%
    filter(sample != "Samp0" & sample != "Samp8" & sample != "Samp9") %>%
    rowwise() %>%
    mutate(pos = list(mk_mat(sample, TRUE)),
           neg = list(mk_mat(sample, FALSE)))
````
